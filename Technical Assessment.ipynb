```python
import random
import time
from datetime import datetime

nodes = ["Ceiling", "Floor", "Wall", "Table"]

print("Simulation started... check your React Dashboard!")

try:
    while True:
        for node_name in nodes:
            value = random.randint(0, 100) 
            status = "Normal"
            if value > 50:
                status = "critical"
            
            data = {
                "node": node_name, 
                "gas_level": value,
                "status": status,
                "time": datetime.now()
            }

            collection.insert_one(data)
            print(f"Sent: {node_name} | Value: {value} | Status: {status}")

        print("--- All nodes updated. Waiting 10 seconds ---")
        time.sleep(10) 
except KeyboardInterrupt:
    print("Simulation stopped.")
```
